```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(4)
```

```python
# class A cluster
A_x1 = np.random.normal(loc=3, scale=0.3, size=10)
A_x2 = np.random.normal(loc=3, scale=0.3, size=10)
```

```python
# class B cluster
B_x1 = np.random.normal(loc=4, scale=0.3, size=10)
B_x2 = np.random.normal(loc=4, scale=0.3, size=10)
```

```python
df = pd.DataFrame({
    'Feature_1': np.concatenate([A_x1,B_x1]),
    'Feature_2': np.concatenate([A_x2, B_x2]),
    'Target': ['Class_A'] * 10 + ['Class_B'] * 10
})
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(
    data=df,
    x='Feature_1',
    y='Feature_2',
    s=100,
    hue='Target'
)
```

```python
sns.scatterplot(
    data=df,
    x='Feature_1',
    y='Feature_2',
    hue='Target',
    s=80
)

plt.scatter(
    x=[3.5],
    y=[3.5],
    c='black',
    s=200,
    marker='x',
    label='new point'
)
plt.legend()
```

```python
data_point = np.array([3.5,3.5])

df['distance'] = np.sqrt(
    (df['Feature_1']-data_point[0])**2 +
    (df['Feature_2']-data_point[1])**2
)
df
```

```python
df.sort_values(by='distance', inplace=True)
df
```

```python
data = pd.read_csv('./Classified Data.txt', index_col=0)
data.head()
```

```python
data.info()
```

```python
features = data.drop(columns='TARGET CLASS')
target = data['TARGET CLASS']

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_fetures = scaler.fit_transform(features)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    scaled_fetures,
    target,
    train_size=.8,
    random_state=42
)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=13)
```

```python
knn.fit(X_train, y_train)
```

```python
predictions = knn.predict(X_test)
predictions
```

```python
from sklearn.metrics import  classification_report
print(classification_report(y_test, predictions))
```

```python
from sklearn.metrics import accuracy_score

k_values = range(1, 50 , 2)
test_scores = []

for i in k_values:
    model= KNeighborsClassifier(n_neighbors=i)
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    acc = accuracy_score(y_test, pred)
    test_scores.append(acc)

max_acc_ind = np.argmax(test_scores)
best_acc_k = k_values[max_acc_ind]


print(f'Best score: {max(test_scores)}')
print(f'Best k value: {best_acc_k}')
```
