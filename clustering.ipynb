```python
from sklearn.datasets import make_blobs 
```

```python
data = make_blobs(
    n_samples=200,
    n_features=2,
    centers=4,
    random_state=42,
)
data
```

```python
import pandas as pd 
df = pd.DataFrame(data=data[0], columns=['feature1', 'feature2'])
```

```python
df['clusters'] = data[1]
df.head()
```

```python
import seaborn as sns 
sns.scatterplot(
    data=df,
    x='feature1',
    y='feature2',
    hue='clusters',
    palette='viridis',
)
```

```python
from sklearn.cluster import KMeans 
kmean = KMeans(n_clusters=4)

kmean.fit(data[0])
```

```python
kmean.cluster_centers_
```

```python
labels = kmean.labels_
```

```python
from matplotlib import pyplot as plt 
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,6))

ax1.set_title("Original")
sns.scatterplot(
    data=df,
    x='feature1',
    y='feature2',
    hue='clusters',
    palette='viridis',
    ax=ax1,
)


ax2.set_title('K means')
sns.scatterplot(
    data=df, 
    x='feature1',
    y='feature2',
    hue=labels,
    palette='viridis',
    ax=ax2,
)
```
