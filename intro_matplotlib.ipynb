```python
y = x*5 + 1
y
```

```python
plt.subplot(1,2,2)
plt.plot(x,y,'r--')
plt.subplot(1,2,1)
plt.plot(x,y)
plt.show()
```

```python
from matplotlib.axes import Axes
fig = plt.figure()
ax:Axes = fig.add_axes([0,0,1,1])


ax.plot(x, x+1 , color="red" , linewidth=0.25)
ax.plot(x,x+2, color='red', linewidth=0.50)
ax.plot(x,x+3, color='red', linewidth=1.00)
ax.plot(x,x+4, color='red', linewidth=2.00)
```

```python
x = np.random.randint(1,10,50)
y = np.random.randint(10,100,50)
color = np.random.randint(10,100,50)
plt.scatter(x,y, marker="*", c=color)
plt.show()
```

```python
brands = ['onePlus', 'Samsung', 'Xiaomi', 'Apple', 'Vivo']
x = [22,35,30,6,10]
c = ['red', 'silver', 'blue', 'pink', 'orange']
ex = [0,0,0,0,0.1]
plt.pie(x, labels=brands, colors=c, explode=ex, shadow=True, autopct = '% .2f')
plt.show()
```
