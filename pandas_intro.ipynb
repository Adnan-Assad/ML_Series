```python
left = pd.DataFrame({
    'key1': ['k0', 'k1', 'k2', 'k3'],
    'A' : np.random.randint(1,10, size=4),
    'B' : np.random.randint(1, 10, size=4),
})

right = pd.DataFrame({
    'key1' : ['k0', 'k2', 'k4', 'k5'],
    'C' : np.random.randint(1, 10, size=4),
    'D' : np.random.randint(1, 10, size=4),
})
```

```python
left
```

```python
right
```

```python
pd.merge(left, right, on='key1', how='inner')
```

```python
pd.merge(left, right, on='key1', how='right')
```
