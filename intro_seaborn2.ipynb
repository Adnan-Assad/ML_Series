```python
import pandas as pd
import numpy as np

# Create sample airline flights data
np.random.seed(42)

airlines = [
    "Emirates", "Qatar Airways", "Singapore Airlines", "Turkish Airlines",
    "Indigo", "Biman Bangladesh", "AirAsia", "Etihad", "Lufthansa", "British Airways"
]

origins = ["Dhaka", "Chittagong", "Sylhet", "Kolkata", "Delhi", "Dubai", "Doha"]
destinations = ["London", "New York", "Dubai", "Doha", "Singapore", "Istanbul", "Bangkok"]

rows = 80  # 80 rows for practice
data = {
    "Flight_ID": [f"FL{1000+i}" for i in range(rows)],
    "Airline": np.random.choice(airlines, rows),
    "Origin": np.random.choice(origins, rows),
    "Destination": np.random.choice(destinations, rows),
    "Passengers": np.random.randint(50, 350, rows),
    "Distance_km": np.random.randint(300, 12000, rows),
    "Price_USD": np.random.randint(120, 1800, rows),
    "Status": np.random.choice(["On Time", "Delayed", "Cancelled"], rows),
}

df = pd.DataFrame(data)

file_path = "/mnt/data/airlines_flights_data.csv"
df.to_csv(file_path, index=False)

file_path
```
